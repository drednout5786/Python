### Шаг 1, 2

```python
import pandas as pd

genres = ['Drama', 'Action', 'Thriller']
import numpy as np

ratings = pd.read_csv('ratings.csv')

movies = pd.read_csv('movies.csv')

# Узнаем, сколько фильмов в датафрейме movies относится к жанру Fantasy
movies['genres'].str.contains('Fantasy').sum()
```

### Шаг 3

```python
joined = ratings.merge(movies, on='movieId', how='left')

crm = pd.read_csv('crm_stats.tsv', sep='\t')
direct = pd.read_csv('direct_stats.tsv', sep='\t')

# Совмещаем датафрейме по дате и названию компани1
join = crm.merge(direct, on=['date', 'campaign'])
# Считаем стоимость заказа
join['order_cost'] = join['cost'] / join['orders']
# Находим значение этого столбца для landings_promo за 1 января 2018
round(join[(join['campaign'] == 'landings_promo') & (join['date'] == '2018-01-01')]['order_cost'])
```

### Шаг 4

```python
genres = ['Drama', 'Action', 'Thriller', 'Comedy', 'Romance', 'War', 'Mystery', 'Crime']

def genres_ratings(row):
    return pd.Series([row['rating'] if genre in row['genres'] else np.NaN for genre in genres])

joined[genres] = joined.apply(genres_ratings, axis=1)

# Определяем, сколько оценок имеет жанр с наименьшим кол-вом оценок
for genre in genres:
    print(genre, len(joined[genre]) - joined[genre].isna().sum())
```

### Домашнее задание

```python
# Создаем список годов
years = [str(i) for i in range(1950, 2011)]

# Создаем функцию
def production_year(row):
    for year in years:
        if year in row['title']:
            return year
    # Если ни один год не найден, возвращаем 1900
    return '1900'

joined['year'] = joined.apply(production_year, axis=1)

# Отвечаем на поставленный вопрос
joined.groupby('year').mean().sort_values('rating', ascending=False)
```
